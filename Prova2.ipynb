Em sistemas de controle, quatro importantes par√¢metros devem ser analisados em sistemas subamortecidos (aqueles em que os polos aparecem em pares conjugados complexos):
<ul>
<li> Tempo de subida (rise-time) - $ùë°_ùëü$: primeiro instante de tempo em que a resposta ao degrau atinge

<li> Tempo de pico - $ùë°_ùëù$: tempo necess√°rio para resposta ao degrau alcan√ßar o primeiro pico m√°ximo;
<li> Sobressinal - $ùëÄ_ùëù$: definido como valor m√°ximo de pico da resposta ao degrau menos o valor da
resposta ao degrau em regime permanente, ou seja:
$$M_p(\%)=\frac{y(t_p)-y(\infty)}{y(\infty)}\times100\%$$
<li> Tempo de acomoda√ß√£o - $ùë°_ùë†$
: tempo necess√°rio para que a resposta ao degrau fique variando em
um faixa de $¬±5\%$ em rela√ß√£o ao valor em regime permanente.
</ul>
Os par√¢metros podem ser visualizados na Figura 3.

A forma can√¥nica de uma fun√ß√£o transfer√™ncia de um sistema de segunda ordem √©:
$$ùêª(ùë†) =\frac{\omega^2_ùëõ}{ùë†^2 + 2\zeta\omega_ùëõ + \omega^2_ùëõ}$$
Se $0 < \zeta < 1$ o sistema √© subamortecido, e para esse caso, o tempo de subida, o tempo de pico, o
sobresinal e o tempo de acomoda√ß√£o pode ser calculados como:
$$t_r = \frac{\pi-arccos(\zeta)}{\omega_d}$$\
$$t_p=\frac{\pi}{\omega_d}$$\
$$M_p(\%)=100exp\left(-\frac{\zeta\pi}{\sqrt{1-\zeta^2}}\right)$$\
$$t_s=\frac{3}{\zeta\omega_n}$$\
sendo $\omega_ùëë = \omega_ùëõ\sqrt{1-\zeta^2}$
denominada de frequ√™ncia natural normalizada.
