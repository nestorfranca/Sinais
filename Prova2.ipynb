Em sistemas de controle, quatro importantes parâmetros devem ser analisados em sistemas subamortecidos (aqueles em que os polos aparecem em pares conjugados complexos):
<ul>
<li> Tempo de subida (rise-time) - $𝑡_𝑟$: primeiro instante de tempo em que a resposta ao degrau atinge
o valor em regime permanente;
<li> Tempo de pico - $𝑡_𝑝$: tempo necessário para resposta ao degrau alcançar o primeiro pico máximo;
<li> Sobressinal - $𝑀_𝑝$: definido como valor máximo de pico da resposta ao degrau menos o valor da
resposta ao degrau em regime permanente, ou seja:
$$M_p(\%)=\frac{y(t_p)-y(\infty)}{y(\infty)}\times100\%$$
<li> Tempo de acomodação - $𝑡_𝑠$
: tempo necessário para que a resposta ao degrau fique variando em
um faixa de $±5\%$ em relação ao valor em regime permanente.
</ul>
Os parâmetros podem ser visualizados na Figura 3.

image.png

A forma canônica de uma função transferência de um sistema de segunda ordem é:
$$𝐻(𝑠) =\frac{\omega^2_𝑛}{𝑠^2 + 2\zeta\omega_𝑛 + \omega^2_𝑛}$$
Se $0 < \zeta < 1$ o sistema é subamortecido, e para esse caso, o tempo de subida, o tempo de pico, o
sobresinal e o tempo de acomodação pode ser calculados como:
$$t_r = \frac{\pi-arccos(\zeta)}{\omega_d}$$\
$$t_p=\frac{\pi}{\omega_d}$$\
$$M_p(\%)=100exp\left(-\frac{\zeta\pi}{\sqrt{1-\zeta^2}}\right)$$\
$$t_s=\frac{3}{\zeta\omega_n}$$\
sendo $\omega_𝑑 = \omega_𝑛\sqrt{1-\zeta^2}$
denominada de frequência natural normalizada.
